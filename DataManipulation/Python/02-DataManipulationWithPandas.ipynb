### pandas.DataFrame()  - 2 Dimensões - 2D

pandas.Panel()           - 3 Dimensões - 3D

numpy.ndarray()     - x Dimensões

```python
import pandas as pd
import numpy as np

stock = {'AMZN': pd.Series([346.15,0.59,459,0.52,589.8,158.88],
                           index = ['Closing price','EPS','Shares Outstanding(M)','Beta', 'P/E','Market Cap(B)']),
         'GOOG': pd.Series([1133.43,36.05,335.83,0.87,31.44,380.64],
                           index = ['Closing price','EPS','Shares Outstanding(M)','Beta','P/E','Market Cap(B)']),
         'FB': pd.Series([61.48,0.59,2450,104.93,150.92],
                           index = ['Closing price','EPS','Shares Outstanding(M)','P/E', 'Market Cap(B)']),
         'YHOO': pd.Series([34.90,1.27,1010,27.48,0.66,35.36],
                           index=['Closing price','EPS','Shares Outstanding(M)','P/E','Beta', 'Market Cap(B)']),
         'TWTR':pd.Series([65.25,-0.3,555.2,36.23],
                           index=['Closing price','EPS','Shares Outstanding(M)','Market Cap(B)']),
         'AAPL':pd.Series([501.53,40.32,892.45,12.44,447.59,0.84],
                           index=['Closing price','EPS','Shares Outstanding(M)','P/E','Market Cap(B)','Beta'])}

print('Dicionário dentro de um DataFrame Pandas:')
stock_df = pd.DataFrame(data=stock)
print(stock_df, '\n')
print(type(stock_df), '\n')

#definindo quais são os índices, quais são colunas e quais são valores
index = ['Closing price', 'P/E']
columns = list(stock.keys())
print('Definindo quais são os índices, quais são colunas(chave) e quais são valores: ')
stock_df1 = pd.DataFrame(data=stock, index = index, columns = columns)
print(stock_df1, '\n')
```

```python
print('Visualizando indices do DataFrame criado: ')
print(stock_df1.index, '\n')
print('Visualizando colunas do DataFrame criado: ')
print(stock_df1.columns, '\n')
```

### Operações com DataFrames

```python
import pandas as pd
import numpy as np

#dicionários aninhados
Quadro_Medalhas_Rio_2016 = {
    'USA' : {'Ouro':46, 'Prata':37, 'Bronze':38},
    'China':{'Ouro':26, 'Prata':18, 'Bronze':26},
    'Britain':{'Ouro':27, 'Prata':23, 'Bronze':17},
    'Russe':{'Ouro':19, 'Prata':18, 'Bronze':19},
    'Germany':{'Ouro':17, 'Prata':10, 'Bronze':15}}

rio_2016 = pd.DataFrame.from_dict(Quadro_Medalhas_Rio_2016)
print('Quadro de medalhas Olimpíadas 2016 RJ:', '\n')
print(rio_2016, '\n', 'type: ', type(rio_2016), '\n')

print('Medalhas da China:')
Medalhas_China = rio_2016['China'] #ou poderia ser chamado como rio_2016['China'] diretamente.
print(Medalhas_China, '\n')

print('Medalhas da Rússia:')
print(rio_2016.Russe, '\n')

print('Medalhas da Alemanha e USA:')
print(rio_2016[['Germany', 'USA']],'\n') #DataFrame[indice[lista das colunas para exibir]]

print('Medalhas da Alemanha:')
print(rio_2016.get('Germany'), '\n')
```

### Slicing

```python
#retornando "fatias" do DataFrame

print('Quadro de medalhas Olimpíadas 2016 RJ:', '\n')
print(rio_2016, '\n', 'type: ', '\n')

print('retorna tudo à partir do indice = 2 (PRATA)')
print(rio_2016[2:], '\n')

print('todas as linhas e colunas onde os valores são a passo 2 (step, pulam)')
print(rio_2016[::2], '\n')

print('todas as linhas e colunas onde os valores são a passo -3 (step, pulam)')
print(rio_2016[::-3], '\n')
```

### Slicing por Indexação
#### DataFram.loc -> retorna o conjunto de dados baseado no valor específico DO ÁLIAS DO ÍNDICE (labels)

```python
print('Quadro de medalhas Olimpíadas 2016 RJ:', '\n')
print(rio_2016, '\n', 'type: ', '\n')

print('Fatiamento por índice [Ouro]')
print(rio_2016.loc['Ouro'], '\n')

print('Fatiamento por índice, considerando apenas [USA]')
print(rio_2016.loc[:,'USA'], '\n')

print('Fatiamento por índice [Prata], considerando apenas [China]')
print(rio_2016.loc['Prata','China'], '\n') # ou DataFrame.loc['linha'],['coluna']

print('Fatiamento por índice [Ouro], considerando verificar se teve mais de 20 medalhas','\n')
print(rio_2016.loc['Ouro'] > 20, '\n')

print('Quais registros tem medalha de [Ouro] acima de 20')
print(rio_2016.loc[:, rio_2016.loc['Ouro'] > 20], '\n')
```

#### DataFram.iloc -> retorna o conjunto de dados baseado no valor específico DA POSIÇÃO DO ÍNDICE (inteiro). Lembrando que a primeira do índice inicia em zero.

```python
#iloc[indice:valor]
print('Quadro de medalhas Olimpíadas 2016 RJ:')
print(rio_2016, '\n', 'type: ', '\n')
print('Fatiamento por posição do índice [1:] -> todas as medalhas de :')
print(rio_2016.iloc[1:], '\n')
print('Fatiamento por posição do índice [2,0:] -> todas as medalha de Prata de todos os países, transformando as CHAVES em ÍNDICES (labels):')
print(rio_2016.iloc[2,0:], '\n')
print('Fatiamento por posição do índice [2:] -> todas as medalha de Prata de todos os países:')
print(rio_2016.iloc[2:], '\n')
print('Fatiamento por posição do índice [:2] -> todas as medalha de Bronze e Ouro de todos os países:')
print(rio_2016.iloc[:2], '\n')
```

```python
print('Fatiamento por posição do índice [:2] -> todas as medalha de Bronze e Ouro de todos os países:')
rio_2016.iloc[:2]
```

# DataFrames

pandas.Series()          - 1 Dimensão  - 1D

pandas.DataFrame()  - 2 Dimensões - 2D

### pandas.Panel()           - 3 Dimensões - 3D

numpy.ndarray()     - x Dimensões

```python
import pandas as pd
import numpy as np

stockData = np.array([[[115.65, 70.3, 88.4],
                       [32.37, 7.65, 99.45],
                       [7.88, 9.22, 24.56]],
                      [[75.65, 6.41, 14.28],
                       [9.30, 2.33, 7.44],
                       [10.44, 25.55, 34.46]]])

print('Estrutura tridimensional em NUMPY do tipo', type(stockData), 'que significa que existem N arrays aninhados')
print('dentro da mesma estrutura.')
stockData
```

### PANEL -> estrutura de dados em CUBO no Pandas que trabalha com apenas 3 dimensões.
#### Havendo a necessidade de manipular mais de 3 dimensões, procurar trabalhar com a função NDARRAY do NUMPY.

```python
stockHistoricalPrices = pd.Panel(data=stockData,
                                 items=('FB', 'NTFLX'),
                                 major_axis = pd.date_range('2/3/2016', periods=3),
                                 minor_axis = ['open price', 'closing price', 'volume'])
stockHistoricalPrices
```

### Criação de um DataFrame a partir de um dicionário

```python
ChinaData = pd.DataFrame(np.array([[1133.68, 390.28],
                                 [1266.83,1198.48],
                                 [1339.72, 9923.47]]),
                       columns=['População(M)','PIB($B)'],
                       index=[1995,2005,2015])

BrasilData = pd.DataFrame(np.array([[115.68, 311.88],
                                   [180.83, 1500.11],
                                   [220.72, 2500.47]]),
                         columns=['População(M)','PIB($B)'],
                         index=[1995,2005,2015])
print('PIB Brasil')
print(BrasilData,'\n')

print('PIB China')
print(BrasilData)
```

```python
print('Unindo os dois DataFrames criados em Pandas, anteriormente [BrasilData + ChinaData], gerando')
print('o DICIONÁRIO Brasil_ChinaData. A união destes dois conjuntos de dados se dá através dos valores das LABELs')
print('dos índices, que no caso são os anos [1995,2005,2015].', '\n')

Brasil_ChinaData = {'Brasil' : BrasilData,
                    'China' : ChinaData,}

print(Brasil_ChinaData, '\n')
print('Type do objeto [Brasil_ChinaData]:', type(Brasil_ChinaData))
```

```python
print('Convertendo o DICIONÁRIO [Brasil_ChinaData], criado anteriormente, para um PANEL em Pandas...', '\n')
df_2 = pd.Panel(Brasil_ChinaData)
print(df_2, '\n')
print('Tipo de dados gerado com a conversão:', type(df_2))
```

### Convertendo DataFrame para Panel
#### MultiIndex -> função para definir diferentes níveis de índice dentro de um determinado objeto

```python
#criando multiindex, onde os níveis serão os nomes dos países (1º nível) e
#as próprias label's dos índices [1995, 2005, 2015] (2º nível)

multiIndex = pd.MultiIndex(levels = [['Brasil', 'China'], [1995, 2005, 2015]],
                     labels = [[1,1,1,0,0,0], [0,1,2,0,1,2]])

#populando um DataFrame em Pandas, usando o multíndice criado anteriomente para os dois níveis: ]Nome dos Paises] e [Ano]
BrasilChina_DF = pd.DataFrame({'População(M)' : [1133.68, 1266.83, 1339.72, 115.68, 180.83, 220.72],
                               'PIB($B)' : [311.88, 2453.11, 5500.47, 390.28, 1500.48, 2500.47]},
                             index = multiIndex)
BrasilChina_DF
```

```python
BrasilChina_DF.to_panel()
```

### Group By e Reshape
#### Sumarizando Dados

```python
import numpy as np
import pandas as pd

#usando encoding='latin-1' porque o aquivo contêm caracteres como ascentuação
cidades_digitais = pd.read_csv(filepath_or_buffer='cidades_digitais.csv', sep=';', encoding='latin-1')
print('Quantidade total de registros por coluna:')
print(cidades_digitais.count(), '\n')
print('Imprimindo as primeiras linhas do DataFrame...')
cidades_digitais.head()
```

```python
print('Removendo a coluna [IBGE]...')
del cidades_digitais['IBGE']

print('Imprimindo as primeiras linhas do DataFrame...')
cidades_digitais.head()
```

```python
print('Aplicando group by ao campo [UF], colocando o resultado em outra variável (BOA PRÁTICA). Evita-se alterar')
print('o DataFrame ORIGINAL:', '\n')
UF_Agrupado = cidades_digitais.groupby('UF')

print('type do novo objeto criado [UF_Agrupado]: ', type(UF_Agrupado))
```

```python
print('Verificando os grupos guardados na variável [UF_Agrupado]:', '\n')
print('A leitura é a seguinte: dentro do estado XX, tenho os elementos dos índices [X, Y, Z, ... N] ')
UF_Agrupado.groups
```

```python
print('Verificando a quantidade de grupos guardados na variável [UF_Agrupado]:', '\n')
print('A leitura é a seguinte: quantas UF tenho dentro do DataFrame [UF_Agrupado]')
len(UF_Agrupado.groups)
```

```python
print('Verificando a quantidade de elementos contidos em cada um dos grupos guardados na variável [UF_Agrupado],')
print('ordenados de forma DESCENDENTE:')
qtd_elementos_UF = UF_Agrupado.size()
qtd_elementos_UF.sort_values(ascending=False)
```

```python
print('Aplicando group by aos campos [UF] e [STATUS], colocando o resultado em outra variável (BOA PRÁTICA). Evita-se alterar')
print('o DataFrame ORIGINAL:', '\n')
UF_STATUS_Agrupado = cidades_digitais.groupby(['UF', 'STATUS'])

print('Verificando a quantidade de elementos contidos em cada um dos grupos guardados na variável [UF_STATUS_Agrupado],')
print('ordenados de forma DESCENDENTE:')
qtd_elementos_UF_STATUS = UF_STATUS_Agrupado.size()
qtd_elementos_UF_STATUS.sort_values(ascending=False)
```

```python
print('Criando um índice para a coluna [STATUS]...', '\n')
STATUS_Indice = cidades_digitais.set_index(['STATUS'])

print('Aplicando group by pelo ÍNDICE = [STATUS], colocando o resultado em outra variável (BOA PRÁTICA). Evita-se alterar')
print('o DataFrame ORIGINAL.', '\n')
print('Mesmo conceito de PIVOT do B.I.')
STATUS_Agrupado = STATUS_Indice.groupby(level = 'STATUS')
STATUS_Agrupado.head()
```

```python
print('Criando um índice para as colunas [STATUS] e [UF]...', '\n')
STATUS_UF_Indice = cidades_digitais.set_index(['STATUS', 'UF'])
STATUS_UF_Indice.head(10)
```

```python
print('Criando um índice para as colunas [UF] e [STATUS]...', '\n')
STATUS_UF_Indice1 = cidades_digitais.set_index(['UF', 'STATUS'])

print('Aplicando group by pelos ÍNDICES = [UF] e [STATUS], colocando o resultado em outra variável (BOA PRÁTICA). Evita-se alterar')
print('o DataFrame ORIGINAL.', '\n')
print('Mesmo conceito de PIVOT do B.I.')
STATUS_UF_Agrupado1 = STATUS_UF_Indice1.groupby(level = ['UF', 'STATUS'])

print('Somando os valores (função SUM do Pandas), agrupando pelo [UF] e [STATUS]...')
STATUS_UF_Agrupado1.sum()
```

```python
print('Somando os valores (função SUM do Numpy), agrupando pelo [UF] e [STATUS]...')
STATUS_UF_Agrupado1.aggregate(np.sum)
```

```python
print('Somando os valores (função SUM do Numpy), agrupando pelo [UF] e [STATUS]...')
STATUS_UF_Agrupado1.aggregate([np.sum, np.mean, np.size])
```

```python
print('Criando índice para as colunas [STATUS, CIDADE, VALOR_TOTAL_PREVISTO, VALOR_INVESTIDO, POPULAÇÃO, PONTOS_ATENDIDOS]...', '\n')
GERAL_Indice1 = cidades_digitais.set_index(['STATUS', 'CIDADE', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO', 'POPULAÇÃO', 'PONTOS_ATENDIDOS'])

print('Aplicando group by pelos ÍNDICES = [STATUS, VALOR_TOTAL_PREVISTO, VALOR_INVESTIDO], colocando o resultado em outra variável (BOA PRÁTICA). Evita-se alterar')
print('o DataFrame ORIGINAL.', '\n')
print('Mesmo conceito de PIVOT do B.I.')
GERAL_Agrupado1 = GERAL_Indice1.groupby(level = ['STATUS', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO'])

cidades_digitais.head(5)
```

```python
print('Somando os valores (função SUM do Pandas), usando o agrupamento de índices')
print('[STATUS, VALOR_TOTAL_PREVISTO, VALOR_INVESTIDO]...')

GERAL_Agrupado1.sum()
```
