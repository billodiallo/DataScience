# Organização de Dados - Python

#### MAP (python built-in) -> função em python (map) que pega os dados de entrada e envia cada um dos elementos de dados para consumo de outra função, gerando outra lista contendo os resultados e com o mesmo tamanho da lista inicial. Retorna um objeto iterable.

#### MAP Reduce (Hadoop & Spark) -> modelo de programação funcional para processamento de dados de forma paralela.

```python
print('MAP function -> Python built-in')
print('Créditos da imagem: http://nasemanadaprova.blogspot.com.br/2015/01/programacao-funcional-python.html')
from IPython.display import Image
Image('function-map_python.png')
```

```python
array1 = range(0,15)
print(' type object: ', type(array1), '\n', 'object: ', array1)
```

```python
def calc_potencia8(num):
    return num ** 8

print('elevando na potencia 8 todos os valores do array1 (SEM map()):')
for i in array1:
    print(calc_potencia8(i))
```

```python
#otimizando o processo, usando a função built-in do python, map()
#map() -> map(chamada da função, elementos que serão passado à função)

print('elevando na potencia 8 todos os valores do array1 (COM map()).')
lista_map01 = map(calc_potencia8,array1)

print('que retorna um objeto iterável ->', lista_map01, '\n')
#print(lista_map01, '\n')

print('colocando os itens do objeto', lista_map01, 'em uma lista:')
print(list(lista_map01))
```

```python
def numero(num):
    return num

array2 = list(map(float,range(5,25)))
print('Lista range(5,26):', '\n', array2, '\n')
#print(' Qtd elementos no array:', array2.count, '\n')

from math import sqrt
print('Raiz quadrada da lista de números indexados (array2[0:12]) do array ORIGINAL:')
print(list(map(numero, array2[0:12])),'\n')
print(list(map(sqrt,array2[0:12])))
```

### Filter -> permite filtrar elementos de uma lista, onde o retorno desta é booleano (True ou False). Ainda retorna os valores da lista com base no resultado do retorno de uma função.

### Resumindo: retorna os elementos onde a expressão é verdadeira.

```python
print('FILTER function -> Python built-in')
print('Créditos da imagem: http://nasemanadaprova.blogspot.com.br/2015/01/programacao-funcional-python.html')
from IPython.display import Image
Image('function-filter_python.png')
```

```python
array3 = range(5,24)
print('array original:')
print(list(array3), '\n')

index1 = []

for i in array3:
    index1.append(array3.index(i))

print('lista de indices (posição) de cada elemento do array:')
print(index1, '\n')

def verificaimpar(num):
    'função retorna true para valores ímpares ou false para valores pares'
    return num % 2 != 0

print('para cada valor do array, verifica se é impar (true ou false) com MAP (Python built-in)')
print(list(map(verificaimpar, array3)), '\n')

print('para cada valor do array, verifica se é impar (true ou false).')
print('Além disso, filtra somente os retornos TRUE (proposta de retorno da função), e retorna o valor da posição do índice')
print('com FILTER (Python built-in)')
print(list(filter(verificaimpar, array3)))
```

```python
# Lista de frases
lista_frases = ['Choveu', 
                'Em SP choveu na manhã de terça-feira. Também choveu em Salvador.', 
                'Após uma tarde ensolarada, choveu em Porto Alegre.', 
                'Não fiz nada, pois choveu hoje pela manhã. Assim como choveu toda semana passada!',
                'A chuva molha as plantas.']

print('lista de palavras:','\n')
print(list(lista_frases), '\n')

def conta_palavra_choveu(frase):
    return frase.count('choveu') >=2

print('para cada valor do array, verifica se a palavra "choveu" é encontrada 2 ou mais vezes, filtrando as incidências')
print('(retornos = TRUE). Para estes registros = TRUE (proposta de retorno da função), retorna o valor da posição')
print('do índice com FILTER (Python built-in)', '\n')
print(list(filter(conta_palavra_choveu, lista_frases)), '\n')

print('para cada valor do array, verifica se é True ou False) com MAP (Python built-in), imprimindo os resultados.')
print(list(map(conta_palavra_choveu, lista_frases)), '\n')
```

### Reduce -> aplica uma função a uma lista de elementos, até reduzir a sequência a um único elemento. Essa função foi removidas das funções built-in do Python, e inserida dentro da biblioteca "functools"

```python
print('REDUCE function')
print('Créditos da imagem: http://nasemanadaprova.blogspot.com.br/2015/01/programacao-funcional-python.html')
from IPython.display import Image
Image('function-Reduce_python.png')
```

```python
from functools import reduce

#LAMBDA é uma função anônima, criada em tempo de execução.
#é aplicada uma função (LAMBDA) a uma lista de elementos, agregando o resultado até que o último elemento seja processado.
#o retorno é justamente a agregação de todas as operações executadas pela função em questão, neste caso, a LAMBDA.

array4 = [1,2,3,-20]
sub = reduce((lambda i, j: i-j), array4)
print('array original: ', array4)
print('operação com base na função REDUCE sobre o array informado -> 1-2 = -1 - 3 = -4 - (-20) = +16', '\n')
print('resultado do REDUCE sobre o array informado:', sub)
```

### List Comprehension -> permite realizar filtros e mapeamentos, semelhante ao filter() e map().

```python
print('LIST COMPREHENSION function -> Python built-in')
print('Créditos da imagem: http://python-3-patterns-idioms-test.readthedocs.io/en/latest/Comprehensions.html')
from IPython.display import Image
Image('list-comprehensions_python.png')
```

```python
array5 = range(5,17)

def calc_potencia8(num):
    return num ** 8

print('array original (array5):')
print(list(array5), '\n')

#list comprehension!
print('---> list comprehension <---', '\n')
print('leitura: para cada item do array5, aplicar a função calc_potencia8, e retornar o resultado em um array.')
print([calc_potencia8(item) for item in array5], '\n')

print('leitura: para cada item do array5, imprimir o próprio valor do item, se este contemplar a condição')
print('de filtragem indicada (valor>=11)')
[item for item in array5 if item >=11]

print('usando list comprehension para converter cada item no array5 para o tipo float')
[float(item) for item in array5]
```

### Funções Lambda -> funções anônimas que são criadas em tempo de execução. Depois da execução, deixa de existir.

```python
print('LAMBDA function -> Python built-in')
print('Créditos da imagem: https://www.slideshare.net/melechi/php-53-part-2-lambda-functions-closures-presentation')
from IPython.display import Image
Image('funtion-lambda_python.png')
```

```python
array6 = range(20,40)

print('array original (array6):')
print(list(array6), '\n')

print('imprimindo resultado da função LAMBDA com base nos valores do array6:')
print(list(map((lambda x: x**2), array6)))

print('imprimindo resultado da função LAMBDA com base nos valores do array7:')
print(list(map((lambda x: x**2), array6)),'\n')

print('array original (array7):')
print(list(filter(lambda a: a % 2 == 0, range(1,20))), '\n')
print('imprimindo resultado da função LAMBDA (a**2) onde o conjunto de dados é um range de números de 1-20 onde')
print('a divisão de cada um desses elementos do range tenha resto zero (array7):')
print(list(map(lambda a: a**2, filter(lambda a: a % 2 == 0, range(1,20)))))
```
