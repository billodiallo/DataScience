### Valores NaN (Not a Number) - valores missing

### Concatenando Series

```python
import pandas as pd
import numpy as np

np.random.seed(587)
Serie4 = pd.Series(np.random.rand(10))

print('imprimindo a Serie4:', '\n')
print(Serie4, '\n\n', 'Qtd. elementos da Serie4 criado:', Serie4.size, '\n')

Serie5 = pd.Series(np.random.rand(10))
print('imprimindo a Serie5:', '\n')
print(Serie5, '\n\n', 'Qtd. elementos da Serie5 criado (pd.concat([Serie4, Serie5])):', Serie5.size, '\n')

Serie_Concat1 = pd.concat([Serie4, Serie5])

print('imprimindo a Serie_Concat1:', '\n')
print(Serie_Concat1, '\n\n', 'Qtd. elementos da Serie_Concat1 criado:', Serie_Concat1.size, '\n')

print('PROBLEMA: se imprimir Serie_Concat1[2], irá imprimir os valores de Serie4[2] e Serie5[2]', '\n')
print(Serie_Concat1[2], '\n')

print('SOLUÇÃO: contar a quantidade de elementos, gerar um range de valores de 0-Qtd elementos contados, e usá-los')
print('como índice na série concatenada [Serie_Concat1]. Veja (range(Serie_Concat1.count())):', '\n')
Serie_Concat1.index = range(Serie_Concat1.count())

print(Serie_Concat1, '\n\n', 'Qtd. elementos da Serie_Concat1 com índices ajustados:', Serie_Concat1.size, '\n')

print('imprimindo os valores a partir de determinados índices [Serie_Concat1.reindex([1,5,9,22,60]]:')
print(Serie_Concat1.reindex([1,5,9,22,60]), '\n')

print('preenchendo c/ valores zeros os elementos onde o valor = NaN dentro dos elementos onde o índice = [1,5,9,22,60]')
print('[Serie_Concat1.reindex([1,5,9,22,60], fill_value=0)]:')
print(Serie_Concat1.reindex([1,5,9,22,60], fill_value=0), '\n')
```

### Tratando valores missing (NaN) no Pandas

```python
print('imprimindo os valores do objeto Serie_Concat1 a partir de determinados índices')
print('[Serie_Concat1.reindex([1,5,9,22,60]]:')
print(Serie_Concat1.reindex([1,5,9,22,60]), '\n')

Serie5 = Serie_Concat1.reindex([1,5,9,22,60])

print('tratando valores NaN -> Serie5.ffill()')
print('FORWARD FILL -> pega o último valor válido, e replica nos valores NaN')
print(Serie5.ffill(), '\n')

print('BACK FILL -> desfaz a operação anterior de FORWARD FILL -> Serie5.bfill()')
print(Serie5.bfill(), '\n')

print('FILL NaN -> preencher valores NaN com um valor pré-determinado -> Serie5.fillna()')
print(Serie5.fillna(55), '\n')
```

### DataFrame - Pandas

```python
import pandas as pd

dicionario_1 = {'chave1': [300, 400, 226, 741],
                'chave2': [100, 244, 710, 3],
                'chave3': [70, 8, 3, 16],
                'chave4': [3, 33, 55, 67],
                'chave5': [99, 5, 15, 23]}

print('criando um DataFrame a partir de uma dict [dicionario_1]... pd.DataFrame(data=dicionario_1)')
df1 = pd.DataFrame(data=dicionario_1)

print('imprimindo o dataframe gerado [df1]:')
print(df1, '\n', 'Qtd. elementos da [df1] criado:', df1.size, '\n', 'Type do objeto df1:', type(df1), '\n')
print('tipos de dados dos elementos das colunas (atributos):', df1.dtypes, '\n')

print('indexação de valores com DataFrame Pandas: retornando todos os valores do atributo df1[chave4] ou df1[[3]]')
print(df1['chave4'], '\n')

print('retornando todos os valores do atributo df1[chave5], filtrando por um índice específico df1[chave4][2]')
print(df1['chave5'][2], '\n')

print('slicing: retornando todos os valores de todas as colunas, da 1ª linha até a 3ª linha (excludente) -> df1[:2]')
print(df1[:2], '\n')

print('retornado quais são os índices do DataFrame gerado [df1] -> df1.index')
print(df1.index, '\n')

print('inserindo uma nova coluna do DataFrame gerado [df1] -> df1.insert')
print('df.insert(posição de índice, nome da nova coluna, array de valores)')
df1.insert(0, 'IDREG', [11, 12, 13, 14])
print(df1, '\n')

print('tranformando os valores da coluna criada [IDREG], em ÍNDICE:')
df2 = pd.DataFrame(df1.set_index('IDREG'))
print(df2, '\n')
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import time

print('Quebrando os valores da coluna "Tempo" do df7 - HH:MM:SS...', '\n')

def calcula_total_segundos(tempo):
    if tempo is np.nan: #testa se valor do parâmetro de entrada = NaN
        return np.nan #retornando "True", ou seja, se é NaN, retorna o próprio valor do parâmetro de entrada (NaN)
    elif str(tempo)=='0':
        return int(tempo)
    else:
        horas, minutos, segundos = str(tempo).split(':') #desmembra o valor do parâmetro de entrada em 3 variáveis: horas, minutos, segundos
        segundos = int(segundos) + (60 * int(minutos)) + (60 * 60 * int(horas)) #calcula a quantidade total em segundos do tempo percorrido
        return int(segundos) #retorna o valor total em segundos

tempo_indice = df7.Tempo[7]
        
print('HH:MM:SS do índice df7.Tempo[7]:', tempo_indice)
print('Qtd. total em segundos do tempo percorrido:', calcula_total_segundos(tempo_indice))
```

```python
df7['Segundos'] = df7['Tempo'].map(calcula_total_segundos)

print('Primeiras 10 linhas do DataFrame [df7]:'); df7.head(10)
```

```python
df7['Minutos'] = df7['Segundos'].map(lambda x: int(x) / 60)
print('Primeiras 10 linhas do DataFrame [df7]:'); df7.head(10)
```

```python
df7['Horas'] = df7['Segundos'].map(lambda x: (int(x) / 60)/60)
print('Primeiras 10 linhas do DataFrame [df7]:'); df7.head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

df8 = pd.read_csv(filepath_or_buffer='df_bicicleta-taxi_backup01.csv', parse_dates=['Data'], usecols=['Data', 'Distancia', 'Tempo', 'Segundos', 'Minutos', 'Horas', 'Minutos_por_KM'])
#df8 = pd.read_csv(filepath_or_buffer='df_bicicleta-taxi_backup01.csv', parse_dates=['Data'], usecols=list(range(0,8)))

print('Convertendo a coluna df8["Segundos"] de tipo de dados int64 para float64...')
df8['Segundos'] = df8['Segundos'].astype('float64')

print('Tipagem das colunas do df8 contruído:')
print(df8.dtypes, '\n')

print('Primeiras 10 linhas do DataFrame [df8]:'); df8.head(10)
```

```python
print('Ordenando DataFrame pelo índice atual [Data]...')
df8.sort_index(inplace=True)
print('Primeiras 10 linhas do DataFrame [df8]:'); df8.head(10)
```

```python
print('Maior tempo de uso das bicicletas (APPLY):', df8.Minutos.max(), 'minutos', '\n')
print('Maior distância de uso das bicicletas (APPLY):', df8.Distancia.max(), 'km', '\n')
print('Menor tempo de uso das bicicletas (APPLY):', df8.Minutos.min(), 'minutos (sujeira!)', '\n')
print('Menor tempo de uso das bicicletas que seja maior que zero minutos (SPLIT + APPLY):', df8.Minutos[df8.Minutos > 0].min(), 'minutos', '\n')
print('Menor distância de uso das bicicletas (APPLY):', df8.Distancia.min(), 'km', '\n')
print('Maior quantidade de utilização das bicicletas onde a distância percorrida seja maior que zero (km) (SPLIT + APPLY):')
print(df8.Distancia[df8.Distancia > 0].value_counts().index[0], 'km','\n')
print('Quantidade de utilização das bicicletas onde a distância percorrida seja maior que zero (km) (SPLIT + APPLY):')
print(df8.Distancia[df8.Distancia > 0].value_counts(), 'km','\n')
```

```python
print('Construindo plots usando as técnicas de SPLIT + APPLY + COMBINE...', '\n')
print('Imprimindo o plot das distâncias de uso das bicicletas, acumuladas ao longo dos dias:')
df8.Distancia.cumsum().plot()
plt.ylabel('Dia')
plt.xlabel('Distância (km)')
```

```python
print('Aplicando o fillna, aos valores NaN da coluna Distância, e ao mesmo tempo, plotando os registros do período:', '\n')
print('Imprimindo o plot das horas de uso das bicicletas, acumuladas ao longo dos dias:')
(df8.Minutos.fillna(0).cumsum()/60).plot()
plt.ylabel('Dia')
plt.xlabel('Horas')
```

```python
print('Extraindo o dia da semana de um valor específico do índice...')
print(df8.Data[0], '-', df8.Data[0].strftime('%A')) # '%A' -> DIA DA SEMANA
```

```python
print('SPLIT + APPLY + COMBINA - CRIAÇÃO de uma nova coluna com base na APLICAÇÃO DE UMA FUNÇÃO que COMBINAVA os dados da coluna DATA X seu respectivo dia da semana (função).', '\n')
print('Criando uma nova coluna de Dia da Semana apoiada pela função strftime do Python:', '\n')
df8['Dia_Semana'] = df8.Data.map(lambda x: x.strftime('%A'))

print('Tipagem das colunas do df8 contruído:')
print(df8.dtypes, '\n')

print('Primeiras 10 linhas do DataFrame [df8]:'); df8.head(10)
```
