## 1. Plotando gráficos com Matplotlib
## 2. Regressão Linear - basicamente é a plotagem de uma linha que identifica a correlação entre duas variáveis, mostrando o quão estão relacionadas.
### 2.1 Fórmula de Regressão Linear: y = a + bx, onde:
### y = variável DEPENDENTE. Valor PREVISTO de y dados um valor para x
### x = variável INDEPENDENTE
### a = ponto onde a linha intercepta o eixo y
### b = inclinaap da linha reta

````python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

preco_casa = np.array([128, 365, 216, 371, 556, 321, 430, 591])
print('Valores dos preços de casas:')
print(preco_casa, '\n')

tamanho_casa = np.array([1140, 1544, 2121, 2255, 2618, 3221, 3605, 4480])
print('Tamanhos das casas:')
print(tamanho_casa, '\n')
````

````python
print('Montando um modelo de Regressão Linear com Scipy...', '\n')
slope, intercept, r_value, p_value, std_err = stats.linregress(preco_casa, tamanho_casa)

print('slope : inclinação da linha de regressão', slope, '\n')
print('intercept : interceptação da linha de regressão', intercept, '\n')
print('r-value : grau de relação entre as variáveis', r_value, '\n')
print('p-value : two-sided p-value for a hypothesis test whose null hypothesis is that the slope is zero.', p_value, '\n')
print('stderr : estimativa do erro padrão', std_err, '\n')
````

````python
print('Construíndo a PRIMEIRA camada do gráfico...(VALORES DO GRÁFICO - VERMELHO)', '\n')
#primeiro parâmetro passa os valores do eixo X
#Segundo parâmetro passa os valores do eixo y
#Terceiro parâmetro passar o estilo do gráfico: 'r' -> cor do marcador em vermelho ||| 'o' -> marcador em círculo
plt.plot(preco_casa, tamanho_casa, 'ro', color='red')
````

````python
print('Construíndo a SEGUNDA camada do gráfico...(LABELS de X e Y)', '\n')

plt.xlabel('Preço da Casa (R$)')
plt.ylabel('Tamanho da Casa (m²)')
````

````python
print('Construíndo a TERCEIRA camada do gráfico...(ESCALAS de X e Y)', '\n')
#Primeiro parâmetro passa os valores para a escala do INÍCIO do eixo X
#Segundo parâmetro passa os valores para a escala do FIM do eixo X
#Terceiro parâmetro passa os valores para a escala do INÍCIO do eixo Y
#Quarto parâmetro passa os valores para a escala do FIM do eixo Y

plt.axis([0, 620, 0, 4600])
````

````python
print('Construíndo a QUARTA camada do gráfico...(VALORES DA REGRESSÃO LINEAR - AZUL)', '\n')

plt.plot(preco_casa, ((preco_casa * slope) + intercept), 'b') # onde 'b' -> blue
````

````python
print('VERMELHO -> valores do cenário real => X = preco_casa e Y = tamanho_casa', '\n')
print('AZUL -> resultado da expressão => (Valor_para_previsão) = ((preco_casa * slope) + intercept).', '\n')
print('Plotando...')

plt.plot()
plt.show()
````

````python
# um gráfico é um conjunto de camadas. Cada camada é um comando que defini-se as regras e definições do gráfico.
# título do gráfico, eixos, escala, figura, tipo de gráfico, legenda, etc.
plt.style.use('classic') #usando estilizaap de plot default

print('Criando uma função para definir a estilização e definições do gráfico...', '\n')

dados_exemplo = np.random.randn(5500,7) #constrói 5000 arrays de 6 elementos

def standard_plot_histogram (array_entrada):
    (figure, axes) = plt.subplots(figsize= (12,5)) #camada da figura e eixos
    (n, bins, patches) = axes.hist  (
                                    array_entrada,
                                    10, #Quantidade de frequências para exibir
                                    normed = 1,
                                    histtype = 'bar', #Estilo do gráfico para histograma
                                    label = ['Color 1', 'Color 2', 'Color 3', 'Color 4', 'Color 5', 'Color 6'] #Legenda
                                    )
    axes.set_title('Histograma para Distribuição Normal', fontsize = 25)
    axes.set_ylabel('Dados', fontsize = 16)
    axes.set_xlabel('Frequência', fontsize = 16)
    axes.legend()
    plt.show()
    
print('Usando a função de plotagem de gráficos personalizada através de FUNÇÃO em PYTHON [standard_plot_histogram] combinada com a estilização "classic" do MATPLOTLIB, utilizando como valores de entrada o array [dados_exemplo].', '\n')
standard_plot_histogram(dados_exemplo)
````

## Normalização de Dados
### Técnica de normalização de Min-max: é muitas vezes conhecida como escala de recursos onde os valores de uma faixa numérica de uma característica de dados, ou seja, uma propriedade, são reduzidos a uma escala entre 0 e 1. 

#### Link para demais técnicas de normalização: https://www.quora.com/What-are-the-best-normalization-techniques-in-data-mining

````python
print('Normalização de dados -> Técnica do MIN-MAX')
print('Créditos da imagem: https://www.quora.com/What-are-the-best-normalization-techniques-in-data-mining')
from IPython.display import Image
Image('min-max-normalization.png')
````

````python
print('Fazendo uma cópia do DataFrame df01 para df01_normalizado...')
df01_normalizado = df01.copy()
print('Renomeando o nome das colunas do DataFrame...')
df01_normalizado.rename(columns={"horsepower":"power"}, inplace=True)
print('Selecionando algumas colunas em específico no df01_normalizado...')

df01_normalizado.drop(['aspiration', 'body', 'bore', 'compression ratio','cylinders', 'doors','drive',
                       'engine location', 'engine size','engine type', 'fuel system','fuel type', 'height',
                       'length', 'peak rpm','stroke','wheel base', 'width', 'Total'], axis=1, inplace=True)
print('Imprimindo as 10 primeiras linhas do DataFrame alterado [df01_normalizado]:')
df01_normalizado.head(10)
````

````python
df01_normalizado.describe()
````

````python
print('Normalização com valores mais ALTOS para as vairiáveis "power", "city mpg" e "highway mpg"')

df01_normalizado['power'] = (df01_normalizado['power'] - df01_normalizado['power'].min()) / (df01_normalizado['power'].max() - df01_normalizado['power'].min())
#prova real posição 0: (102 - 48) / (200 - 48) => 0,355263

df01_normalizado['city mpg'] = (df01_normalizado['city mpg'] - df01_normalizado['city mpg'].min()) / (df01_normalizado['city mpg'].max() - df01_normalizado['city mpg'].min())
df01_normalizado['highway mpg'] = (df01_normalizado['highway mpg'] - df01_normalizado['highway mpg'].min()) / (df01_normalizado['highway mpg'].max() - df01_normalizado['highway mpg'].min())
````

````python
print('Verificando ajustes da NORMALIZAÇÃO...(valores maiores)')
print('Imprimindo as 5 primeiras linhas do DataFrame alterado [df01_normalizado]:')
df01_normalizado.head()
````

````python
print('Normalização com valores mais BAIXOS para as variáveis "price", "weight", "riskiness" e "losses"')

df01_normalizado['price'] = (df01_normalizado['price'] - df01_normalizado['price'].min()) / (df01_normalizado['price'].max() - df01_normalizado['price'].min())
#prova real posição 0: 1 - (( - ) / ( - )) => 

df01_normalizado['weight'] = (df01_normalizado['weight'] - df01_normalizado['weight'].min()) / (df01_normalizado['weight'].max() - df01_normalizado['weight'].min())
df01_normalizado['riskiness'] = (df01_normalizado['riskiness'] - df01_normalizado['riskiness'].min()) / (df01_normalizado['riskiness'].max() - df01_normalizado['riskiness'].min())
df01_normalizado['losses'] = (df01_normalizado['losses'] - df01_normalizado['losses'].min()) / (df01_normalizado['losses'].max() - df01_normalizado['losses'].min())
````

````python
import numpy as np
import pandas as pd

df02 = df01.copy()
print('Renomeando o nome das colunas do DataFrame...', '\n')
print('Selecionando algumas colunas em específico no df02...', '\n')
df02.drop(['aspiration', 'body', 'bore', 'compression ratio','cylinders', 'doors','drive',
           'engine location', 'engine size','engine type', 'fuel system','fuel type', 'height',
           'length', 'peak rpm','stroke','wheel base', 'width', 'Total',
           'riskiness', 'losses', 'weight', 'horsepower', 'city mpg', 'highway mpg'
          ], axis=1, inplace=True)


print('Capturando menores valores agrupados por fabricante...', '\n')
min_price = df02.groupby('make').min()
min_price.rename(columns = {min_price.columns[0] : 'min_price'}, inplace = True)

print('Capturando média dos valores agrupados por fabricante...', '\n')
mean_price = df02.groupby('make').mean()
mean_price.rename(columns = {mean_price.columns[0] : 'mean_price'}, inplace = True)

print('Capturando maiores valores agrupados por fabricante...', '\n')
max_price = df02.groupby('make').max()
max_price.rename(columns = {max_price.columns[0] : 'max_price'}, inplace = True)

print('Concatenando min_price, mean_price, max_price por colunas...', '\n')
print('Imprimindo valores do DataFrame criado [df02_new]...')
df02_new = pd.concat([min_price, mean_price, max_price], axis=1)
df02_new
````

### Dashboard - Plots combinados em uma área
#### 1º) Definição das áreas de plotagem dos gráficos

```
--------------------------------------------
|                  título                  |
--------------------------------------------
|          preços (min, médio, max)        |
--------------------------------------------
| gráfico_01          |                    |
|                     |                    |
-----------------------      gráfico_02    |
|gráfico_03|gráfico_04|                    |
--------------------------------------------
|                gráfico_05                |
--------------------------------------------
```

````python
import matplotlib as mpl
import matplotlib.pyplot as plt

# Construindo as camadas (sem os dados)
figure = plt.figure(figsize=(10, 8))
area_plot = mpl.gridspec.GridSpec(4, 2, height_ratios=[1, 2, 8, 2])

# Camada 1 - título
title = mpl.gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec=area_plot[0, :])
title_axes = figure.add_subplot(title[0])

# Camada 2 - preços (min, médio, max)
price = mpl.gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec=area_plot[1, :])
price_axes = figure.add_subplot(price[0])

# Camada 3 - gráfico_01, gráfico_02, gráfico_03, gráfico_04
area_central = mpl.gridspec.GridSpecFromSubplotSpec(2, 2, height_ratios=[2, 1], subplot_spec=area_plot[2, :1])
graphic_01 = figure.add_subplot(area_central[0, :])
graphic_02 = figure.add_subplot(area_central[1, :1])
graphic_03 = figure.add_subplot(area_central[1:, 1])
area_central_dir = mpl.gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec=area_plot[2, 1])
graphic_04 = figure.add_subplot(area_central_dir[0])

# Camada 4 - gráfico_05
area_rodape = mpl.gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec=area_plot[3, :])
graphic_05 = figure.add_subplot(area_rodape[0])

# Une as camas, ainda sem dados
area_plot.tight_layout(figure)
plt.show()
````
